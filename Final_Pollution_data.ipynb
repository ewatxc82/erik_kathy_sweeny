```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
# read pollution csv file
file_path_pollution = ("Resources/pollution_data.csv")


# create the data frame for pollution
pollution_df = pd.read_csv(file_path_pollution)
pollution_df.head()
```

```python
# select the columns
pollution_df = pollution_df[["State", "City", "Date Local", "NO2 Mean", "NO2 AQI", "O3 Units", "O3 Mean", "O3 AQI", "SO2 Mean", "SO2 AQI", "CO Mean", "CO AQI"]]

# display the new dataframe
pollution_df.head()
```

```python
pollution_df['Date Local'] = pd.to_datetime(pollution_df['Date Local'], format= '%m/%d/%Y')
pollution_df['Year'] = pollution_df['Date Local'].dt.strftime('%Y')
pollution_df
```

```python
# change NaN values to 0
pollution_df.fillna(0, inplace=True)
pollution_df.head()
```

```python
# Make DataFrame Specific to Los Angeles
LAdf = pollution_df[pollution_df['City'] == 'Los Angeles']
LAdf.reset_index(drop=True, inplace=True)
LAdf
```

```python
# Group DataFrame By Year
LaAnnual = LAdf.groupby('Year')
LaAnnualAvg = LaAnnual.mean()
LaAnnualAvg
```

```python
# Make DataFrame Specific to New York City
NyDf = pollution_df[pollution_df['City'] == 'New York']
NyDf.reset_index(drop=True, inplace=True)
NyDf
```

```python
# Group DataFrame By Year
NyAnnual = NyDf.groupby('Year')
NyAnnualAvg = NyAnnual.mean()
NyAnnualAvg
```

```python
# read populatin csv file
file_path_population = "Resources/population_data.csv"

## create the data frame for population
population_df = pd.read_csv(file_path_population, encoding = "ISO-8859-1", low_memory=False)
population_df.head()
```

```python
# select the columns
population_df = population_df[["NAME", "STNAME", "POPESTIMATE2000", "POPESTIMATE2001", "POPESTIMATE2002", "POPESTIMATE2003", "POPESTIMATE2004", "POPESTIMATE2005", "POPESTIMATE2006", "POPESTIMATE2007", "POPESTIMATE2008", "POPESTIMATE2009", "POPESTIMATE2010"]]

population_df.head()
```

```python
LA_city= population_df[population_df['NAME'] == 'Los Angeles city']
LA_city
```

```python
NYC = population_df[population_df['NAME'] == 'New York city']
NYC
```

```python
# Now transform the rows of desired data into data frames that can be joined with Pollution Data
# Note: This is not the only way to accomplish this goal.
# Alternatively, the headers could have been renamed to simply the name of the years in a string form then the table could be pivoted then merged. 

# Los Angeles Population DataFrame
LA_city_df = pd.DataFrame({"Year": ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010"],
                       "Population": [3701062, 3733328, 3762432, 3784279, 3792650, 3788197, 3767960, 3751872, 3763566, 3781938, 3792661]})
LA_city_df.set_index('Year', inplace=True)
LA_city_df
```

```python
# New York City Population DataFrame to be joined
NY_df = pd.DataFrame({"Year": ["2000", "2001", "2002", "2003", "2004", "2005", "2006", "2007", "2008", "2009", "2010"],
                       "Population": [8017608, 8059813, 8072000, 8068073, 8043366, 8013368, 7993906, 8013775, 8068195, 8131574, 8175133]})
#NyPopDf.set_index('Year', inplace=True)
NY_df.set_index('Year', inplace=True)
NY_df
```

```python
# Joining these tables 
LAX_df = pd.concat([LaAnnualAvg, LA_city_df], axis=1, join="outer")
LAX_df.head()

# Exporting LA Data Frame to a csv file to be imported into database
LAX_df.to_csv("Resources/LAX_df.csv", index=True, header=True)
```

```python
LAX_df.head()
```

```python
#New York DataFrame to be Loaded into database

NYC_df = pd.concat([NyAnnualAvg, NY_df], axis=1, join="outer")
NYC_df

# Exporting NYC Data Frame to a csv file to be imported into database
NYC_df.to_csv("Resources/NYC_df.csv", index=True, header=True)
```
